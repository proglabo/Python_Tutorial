## クラス

-------------

1. **関数の集まり**

2. **クラスは同じ機能を引き継いだインスタンスというクローン人間みたいなものを生成できる**

```python
"""基本形ー関数
クラスを宣言し、その中に関数を書いていきます
func関数はTestクラスの持ち物です
"""
class Test():
    def func():
        print("クラスの持ち物です")
#これは呼び出せない
func()
```

```python
"""基本形ー関数
クラスを宣言し、その中に関数を書いていきます
func関数はTestクラスの持ち物です
"""
class Test():
    def func():
        print("クラスの持ち物です")
Test.func() #クラス関数を呼び出し
```

## インスタンス

----

1. クラスから作られたクローン
2. 一つのクラスからいくつでも作ることができる

```python
"""インスタンス作成（クローン）
インスタンス名　＝　クラス名（）
でインスタンスを作成できる
"""
class Test():
    def func():
        print("クラスの持ち物です")
test_instance = Test()
```

```python
"""インスタンスの利用（失敗例）
インスタンスはクラスと同様の機能を引き継ぐ
"""
class Test():
    def func():
        print("クラスの持ち物です")
test_instance = Test()
test_instance.func()
```

```python
"""インスタンスの利用（成功例）
インスタンスはクラスと同様の機能を引き継ぐが、
何を引き継ぐかはこちらが定義してやらないといけない
"""
class Test():
    def func(self):
        print("インスタンスの持ち物です")
test_instance = Test()
test_instance.func()
```

## 用語の定義

**クラス関数**：クラスだけがアクセスできる関数

**クラス変数**：クラスだけがアクセスできる変数

**メンバ関数**：インスタンスのみがアクセスできる関数

**メンバ変数**：インスタンスのみがアクセスできる変数
