## 関数
一連の処理をグループ化し、まとめて実行するものを関数といいます

```python
"""通常の関数
関数名は自由に決めれます
"""
def func():
    x=1
    y=x+1
    print(y)
func()
```

```python
"""引数を持った関数
引数：関数が受け取る値
例：「y=x+1のxに2を代入しなさい」という問題の2に該当する
"""
def func(hikisuu):
    x=hikisuu
    y=x+1
    print(y)
func(2)
```

```python
"""引数を持った関数
どんな型の引数でも大丈夫
"""
def func(hikisuu):
    print(hikisuu)
func("文字列型の引数")
```

```python
"""複数の引数を持った関数
"""
def func(hikisuu1,hikisuu2):
    y=hikisuu1*hikisuu2
    print(y)
func(2,3)
```

```python
"""返り値のある関数
返り値：関数の出力
関数の実行結果を変数に保存することができます
"""
def func():
    return "関数の返り値"
text=func()
print(text)
```

```python
"""返り値と引数のある関数
"""
def func(x):
    y=x+1
    return y
y=func(1)
print(y)
```

```python
"""ネストされた関数
関数の中に関数を含むこともできます
"""
def func():
    def func_in_func():
        print("ネストされた関数が実行されました")
    print("func1が実行されました")
    func_in_func()
func()
```

```python
"""再帰的に呼び出す
関数をその関数自身で呼び出すこともできます
"""
def func():
    print("funcを呼び出しました")
    func()
func()
```

```python
"""可変長引数（タプル）
引数は通常一対一対応ですが、「*」をつけることで
タプル型で自由な長さの引数を渡せます
"""
def func(*arg):
    print(arg)
    print(type(arg))
func(1,2,3,34)
```

```python
"""可変長引数（辞書）
引数は通常一対一対応ですが、「**引数」とすることで
辞書型で自由な長さの引数を渡せます
"""
def func(**keywords):
    print(keywords)
    print(type(keywords))
func(x=1)
func(x1=1,x2=2)
```
